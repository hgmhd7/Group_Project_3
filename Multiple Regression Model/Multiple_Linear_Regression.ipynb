```python
import pandas as pd
from sklearn import linear_model
import statsmodels.api as sm
```

```python
clean_wine_df = pd.read_csv('../data/clean_wine.csv')
```

```python
clean_wine_df.head()
```

```python
clean_wine_df = clean_wine_df.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis=1)
```

```python
clean_wine_df.head()
```

```python
clean_wine_df.columns[clean_wine_df.isna().any()].tolist()
```

```python
no_nans_wine_df = clean_wine_df.dropna(subset = ['price', 'freq', 'red', 'rose', 'white', 'sparkling'])
no_nans_wine_df.head()
```

```python
X = no_nans_wine_df[['price', 'freq', 'red', 'rose', 'white', 'sparkling']]
Y = no_nans_wine_df['points']
```

```python
regr = linear_model.LinearRegression()
regr.fit(X, Y)
```

```python
print('Intercept: \n', regr.intercept_)
print('Coefficients: \n', regr.coef_)
```
